```python
import geopy
from geopy.geocoders import Nominatim

def get_location_from_coordinates(latitude, longitude):
    """
    Retrieves location information (address, city, state, country)
    from given latitude and longitude coordinates.

    Args:
        latitude (str or float): The latitude coordinate.
        longitude (str or float): The longitude coordinate.

    Returns:
        dict: A dictionary containing address details, or None if not found.
    """
    geolocator = Nominatim(user_agent="location_retriever")
    location = geolocator.reverse((latitude, longitude), exactly_one=True)

    if location:
        address = location.raw['address']
        return {
            'address': address,
            'latitude': location.latitude,
            'longitude': location.longitude,
            'city': address.get('city', ''),
            'state': address.get('state', ''),
            'country': address.get('country', ''),
            'zip_code': address.get('postcode', '')
        }
    else:
        return None

# Example usage:
latitude = "40.7128"  # Example latitude (New York City)
longitude = "-74.0060" # Example longitude (New York City)

location_data = get_location_from_coordinates(latitude, longitude)

if location_data:
    print(f"Location data for {latitude}, {longitude}:")
    print(f"  Address: {location_data['address']}")
    print(f"  Latitude: {location_data['latitude']}")
    print(f"  Longitude: {location_data['longitude']}")
    print(f"  City: {location_data['city']}")
    print(f"  State: {location_data['state']}")
    print(f"  Country: {location_data['country']}")
    print(f"  Zip Code: {location_data['zip_code']}")
else:
    print(f"Could not retrieve location data for {latitude}, {longitude}")
```
